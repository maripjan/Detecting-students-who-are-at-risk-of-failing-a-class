For more comprehensive information, including full data access, detailed data description and methodology for data selection, please visit Ope University's dedicated webpage: https://analyse.kmi.open.ac.uk/open_dataset


</br>

### Evaluation Strategy


Choice of any evluation metric always stems from an objective a problem we re trying to solve. 

In our particular case, the objective is to detect who is likely to fail the course, so that tutors can be allocated to help those students to pass the course. Yet, there is an additional constraint. In particular, due to resource constraints, the number of tutors that can be allocated for such missions are oftentimes relatively small. This is espceially true if class size is large (ex: first-year courses). Thus, curriculum administrators often face the situation where it is not feasible to help everyone who are prone to failing. In those scenarios, costs of both  False Positives and False Negaives are high. Hence, it seems that for those types of problems *F1 score* is an appropriate KPI. A good F1 score implies that the model has been able to achieve both low false positives and low false negatives. In our case this implies that the model was able to correctly identify most of the students who are at risk of failing, while not raising too much false alarms at the same time.

On the other hand, private universities or some highly-speicalized courses tend to have very small class size relative to the number of tutors. Given that availability of tutors is not a big concern, one of the most important questions one should ask is : "There are **N** students who acutally failed. How many of them were identified? This means that the cost of missing an unsuccessful student(*i.e. wrongly identifying him or her as the one who will successfully complete the course*) is significantly higher than mistakenly classifying a successful student as unsuccessful.

In other words, we are better of having more false positives and fewer false negatives rather than other way around. Therefore, although scores for all three metrics will be provided, a special attention should be allocated to recall when comparing models with each other.


Hence,in order to evaluate performance of prediction algorithms, I will use two different metrics: *F1 score* and *Recall*.

#### Overall information about students

```python
student_registration = pd.read_csv('studentRegistration.csv')
student_registration.dropna(inplace = True)
student_registration['date_unregistration'].plot(kind = 'hist', edgecolor='black', bins = 40)
plt.xlabel('Unenrollment before / after start of a course (day 0)')
plt.show()
```

For further predictions, I will only use data for those students who either **passed** or **failed** the course. That is, I will exclude data for those who withdrew from the course, since in my case this is the best strategy to follow. Here is why. 

As you can see from the histogram above, students tend to withdraw from the course at any time before or after course starts (negative numbers indicate that students unenrolled before online course has started). Although technically speaking withdrawal from a course can be considered as a failure,yet merging students who withdrew from a course together with those who failed the class is not an optimal solution for several reasons.

<ol>
  <li>First, it brings an issue of data leakege. That is, we will find ourselves in a situation where we make predictions for someone who might have already unenrolled from the course. 
    </li>
    
    
  <li>Secondly, students can unenroll from a course for a variety of reasons. Therefore, inlcuding them into a prediction model is likely to bring additional noise, which in turn might impact model's predictive power.
    </li>
</ol>    
On the other hand, considering withdrawn students as a separate class bring brings exacctly the same problems as above. What is the point of finding out that a student is likely to withdraw from a class by the time course is about to finish if he or she already did so at the very beginning?


So, once we exclude students who unenrolled from a course and merge together two types of successful students, this is what we obtain.

```python
student_info = pd.read_csv("studentInfo.csv")
student_info = student_info[student_info['final_result'] != 'Withdrawn']

## Since those who completed a course with distinction have to be the ones who passed it.
student_info.loc[student_info['final_result'] == 'Distinction','final_result'] = 'Pass'

student_info['final_result'].value_counts().plot(kind = 'pie', autopct ='%1.0f%%')
plt.show()

student_info.head(3)
```

#### Information about students' online activity

Now, we import file that contains information about students' interactions with a Virtual Learning Enviroment (measured by the number of cliks).  

After importing file, we can find the total number of clicks made by a student enrolled in a particular class that was offered at a particular semester. This will be helpful to see if more interaction with the online learning material (VLE) is associated with higher chances of passing an exam.

```python
student_vle_interact = pd.read_csv('studentVle.csv', usecols=['id_site','code_module','code_presentation','id_student','sum_click']) 
vle = pd.read_csv('vle.csv', usecols = ['id_site', 'code_module', 'code_presentation', 'activity_type'])


## Merge two tables together in order to obtain the complete information regarding student's interaction with learning material
student_interaction = pd.merge(student_vle_interact, vle, on = ['id_site', 'code_module', 'code_presentation'])
```

</br>

After inspecting table, it seems like a good idea to aggregate dataframe accross every student, course, type of activity and semester it was offered in. The advantage of doing so is that it allows us to calculate how many times did a student interact with each kind of learning material in a given course. This data in turn might be helpful in deriving insights  about the importance of each type of learning material (ex: book, video, exercises etc.) in determining success of a student in a given course.

The best way to achieve our intended goal is to encode each activity type as a seperate column header (*a.k.a. pivoting*), because this will create new features, which might be useful.

```python
## Pivot the merged table
student_interaction = pd.pivot_table(data =student_interaction,
                                    values = 'sum_click',
                                    index = ['code_module', 'code_presentation','id_student'],
                                    columns = 'activity_type',
                                    fill_value = 0,
                                    aggfunc=np.sum
                                    )

### Create a new column that represents total number of clicks student has made during the course.
student_interaction['sum_of_clicks'] = student_interaction.loc[:, 'dataplus' : 'url'].sum(axis = 1)
```

</br>

Since the dataset contains information about all clicks made by a student, there is a lot of noise. Namely, some of the variables like folder, homepage and glossary have no practical meaning, since the do not accurately capture the type of material a student has interacted with. Therefore, we can drop them without any significant loss of informativity.


```python
## Drop uninformative columns
student_interaction.drop(columns=['folder', 'glossary', 'homepage'], axis = 1, inplace = True)

### Reset the index to restore features back again.
student_interaction = student_interaction.reset_index().rename_axis(None, axis=1)

student_interaction.head()
```

```python
student_interaction.shape
```

</br>

Note that both **student_interaction** and **student_info** tables have 3 columns in common. Thus, let us join them in order to create a single table. 

```python
student_data = pd.merge(student_info, student_interaction, 
               on=['code_module','code_presentation','id_student'],
               how='left')

print(student_data.shape)
```

</br>

#### Information about course assignments

Lastly, let's upload  files that contain information about students' performances on a subject he or she is enrolled. Since we are interested in predicting whether a student will fail/pass the course, we need to exclude rows that contain information about scores in **Exam**. Otherwise, there is no point of predicting student's performance if the course is over and we know the exam score. However, I will leave scores for otehr two types of assesments, which are likely assignments like homeworks and quizzes.

```python
## Upload data about the type of assessments
assessments = pd.read_csv('assessments.csv')

## Drop rows that contain assessment type Exam
assessments = assessments[assessments["assessment_type"]!="Exam"]

## Upload table about student's performance in each assignment
student_assesment = pd.read_csv('studentAssessment.csv')

## Merge two tables
student_scores = pd.merge(assessments, student_assesment, on = 'id_assessment')
```

</br>

As in the previous case, I will aggreagte data and try to obtain some new features based on existing ones.

```python
### Calculate how often did students miss deadline for submitting assignments.
student_scores['submitted_with_delay'] = np.where(student_scores['date_submitted'] > student_scores['date'], 1, 0)


## At this moment we can get rid of unnecessary columns that are very unlikley to contain any useful information after this point.
student_scores.drop(columns = ['id_assessment', 'date', 'date_submitted', 'assessment_type', 'is_banked'], axis = 1, inplace = True)

###  Impute NaN values with an average score of a student within the subject
student_scores['score'] = student_scores.groupby(['code_module', 'id_student'])['score'].apply(lambda x: x.fillna(x.mean()))

### Even after what I did above, there are still 21 NaN for the score column. To overcome it, I will simply replace them with class average.
student_scores['score'] = student_scores.groupby(['code_module'])['score'].apply(lambda x: x.fillna(x.mean()))


## Create a column that represents a weighted score in an assignment
student_scores['weighted_score'] = (student_scores['weight']/100)*student_scores['score']
```

</br>

Since some assignments have the weight of zero (most likley these are self-study or additional exercises), it makes sense to calculate two different types of total scores for each student. One representing the weighted score, while another one being a total number of points accumulated. The same can also be applied when aclculating a final average score (weithted vs. unweighted) a student has earned during the course.

```python
## Create a column representing performance in non-credit assignments(i.e. with weight of 0)
student_scores['score_in_non-credit_assignments'] = np.where(student_scores['weight']==0, student_scores['score'], 0)


### Aggregate data accross every student, class and semster to compute each student's overall performance.
student_scores = student_scores.groupby(['code_module','code_presentation','id_student']).agg({'score' : ['sum', 'mean'],
                                                                            'submitted_with_delay': ['count'],
                                                                            'weighted_score': ['sum', 'mean'],
                                                                            'score_in_non-credit_assignments':['sum','mean']
                                                                                }).reset_index().round(2)

## Rename column so that they make more sense
student_scores.columns = ['code_module', 'code_presentation', 'id_student', 'unweighted_score', 
                          'unweighted_avg_score', 'times_overdue', 'weighted_score', 'weighted_avg_score',
                          'non_credit_score_total', 'non_credit_score_avg']


display(student_scores.head())
```

</br>

```python
data = pd.merge(student_data, student_scores, 
               on=['code_module','code_presentation','id_student'],
               how='left')

## Additionally, let us delete possibly duplicate rows
data.drop_duplicates(inplace = True)
display(data.head(3))
```

### Initial Data Preprocessing


As a first step before exploring predictive models, let's explore in detail what our table consists of. 

```python
data.info()
```

</br>

Indeed, certain proprocessing procedures are necessary for this dataset.

 1. First, we need to convert some **object** variables into **categorical** features.

 2. Missing values have to be handled. After looking into the dataset and thinking about what could be the best strategy to impute missing data, I arrived to the conclusion that the most logical step would be to replace missing values for all **numeric variables with their respective averages**. Similarly, **categorical variables should be replaced by their most frequent values**.

Before making any adjustments to the dataset, let's copy it and do adjustments with its copy instead of an original data.

```python
df = data.copy()
```

```python
#### Let's change encodings in some columns for a better convenience and interpretability

## Convert imd_band from categorical variable into numerical one
df['imd_band'].replace({'90-100%':10, '20-30%':3, '30-40%':4, '50-60%':6,
                        '80-90%':9, '70-80%':8, '60-70%':7, '40-50%':5,
                        '10-20':2, '0-10%':1}, inplace = True)

df['imd_band'].fillna(df['imd_band'].mean(), inplace = True)

## Encode those with age of over 55 as 55+ 
df['age_band'].replace(to_replace = '55<=', value = '55+', inplace = True)
```

Since for some students in a joint table some course-related information is still misssing, the main concern is what to replace them with. After inspecting the table and thinking about the issue, I decided that such information is likely to be missing because those students never interacted with VLE and never submitted any assignments. This conclusion is reinforced with the fact that almost all of them eventually failed the course. Hence, the most reasonable way would be to substitute those missing fields with zero.

```python
df.fillna(0, axis=1, inplace=True)

## Since ID Number of students is unrelated with their perfromance, we can drop this feature
df.drop('id_student', axis = 1 , inplace = True)
```

```python
df.info()
```

</br>

Now I will encode several columns as categorical variables. Otherwise not all ML algorithms will be able to work with the data in its default representation.

```python
categorical_columns = ['code_module','code_presentation','gender','region','highest_education','age_band','disability']
target = ['final_result']
numeric_columns = [column for column in df.columns if column not in categorical_columns and column not in target]

## Since our main goal is to predict those who are likely to fail, let's change variable encoding in target variable as well
df['final_result'] = df['final_result'].map({'Fail' : 1, 'Pass': 0}).astype(int)

#df = pd.get_dummies(data=df,columns = categorical_columns, drop_first= True)
df.head(3)
```

#### ML Workflow for Maximizing F1 score

</br>

Remember that I mentioned above that university administration or course instructors should select one of the two metrics based on the teaching resources (relative class size) they can devote for tutoring students who are at risk of failing the course. <u>In this fisrt setup, I will focus my attention on those education institutions or courses where number of tutors (relative to a class size) is scarce</u> (which is the most realistic scenario, due to budget constraints)

##### **Data Transforming and  Pre-Modeling Setup**


Before starting the modeling process, I will run **PyCaret's** <code>setup()</code> function in order to split the data in to **Train/Test datasets**, to Rescale and Normalize Data and instruct predictive models I am going to use soon to implement scikit-learn's **Repeated Stratified K-Fold Cross-Validation**

```python
df_setup_no_interaction = setup(data = data_for_modeling,
                 target = 'final_result',    
                 train_size = 0.3,               ### 30% of data will be used for testing
                 numeric_features = numeric_columns,  
                 categorical_features = categorical_columns,
                 fold = 5,    
                 fold_strategy = RepeatedStratifiedKFold(n_splits = 5, n_repeats = 3), ## from scikit-learn's library
                 fold_shuffle = False,
                 normalize = True,   
                 transformation_method = "yeo-johnson",  ## normallize data
                 ignore_low_variance = True,
                 remove_multicollinearity = True,
                 multicollinearity_threshold = 0.95,
                 create_clusters = True,                 
                 use_gpu = True,
                 silent = True, 
                 session_id = 123,
                 verbose = False) ### Do not print the output
```

The <code>compare_models()</code> function below allows us to compare many models at once. This is one of the major advantages of using **PyCaret** as a starting point in any ML workflow. A signle line of code will show you a comparison table among different classification models evaluated under 6 different metrics.  Since I used default parameters, in my particular case <code>compare_models()</code> has trained and evaluated 18 different classification models using Repeated Stratified cross-validation and at the end displayed **15** models that performed the best.

```python
models_comparison = compare_models(sort = 'F1', round = 3, n_select = 5)
```

```python
## Save Gradient Boosting 
gbc = models_comparison[0]
```

#### Hyperparameter Tuning

Now, with an intent to obtain even higher F1 scores, I will tune hyperparameters of several ML algorithms from the table above that have both:

  <ol>
     <li> Demonstrated high scores </li>        
      <li> Demanded relatively low computational Resources</li>
   </ol> 
   
After insepcting the comparison table from above, I came to the conclusion that there are 3 algorithms that simultaneously suit these two creiteria: **Logistic Classifier, Random Forest** and **Light GBM**. Therefore, I will tune them  all in order to explore if tuning can bring any additional value.

##### Tuning Logistic Classifier

```python
lr = create_model('lr', verbose=False)

params = {'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
          'penalty' : ['l1', 'l2'],
          'C' : [100, 10, 1.0, 0.1, 0.01, 0.001],
          'class_weight': [None, 'balanced']
          }

lr_tuned = tune_model(lr, 
                      round = 3,
                      n_iter = 50,
                      custom_grid = params,
                      optimize = 'F1',
                      search_algorithm = 'random',
                      early_stopping = True,
                      choose_better = True)
```

</br>

##### Tuning Random Forest Classifier

```python
rf_tuned = tune_model(create_model('rf', verbose = False), 
                      n_iter = 50,                      
                      optimize = 'F1',
                      search_algorithm = 'random',
                      early_stopping = True,
                      choose_better = True)
```

##### Tuning LightGBM

```python
lgbm = create_model('lightgbm', verbose = False)

params = {
    'learning_rate': [0.01, 0.1, 0.2], # eta-arameter (smaller number makes model more robust by shrinking weights during each step)
    'max_depth': [20, 40, 80], #max depth of a tree
    'min_child_weight': [40, 60, 80], # minimum sum of weights of all observations required in a child (higher values reduce over-fitting)
    'n_estimators': [10, 20, 50, 100],
}

lgbm_tuned = tune_model(lgbm, 
                      n_iter = 50,
                      custom_grid = params,
                      optimize = 'F1',
                      search_algorithm = 'random',
                      early_stopping = True,
                      choose_better = True)
```

```python
print(lgbm_tuned)
```

</br>
It turns out that Hyperparameter Tuning could only marginally increase F1 score for all of the three classifiers considered.

#### Ensembling Models (Blending)

Finally, let's explore if ensembling multiple models together can boost F1 score. The idea behind ensembling is to combine different machine learning algorithms to predict the final outcome. Although downsides of such method are increase in computational cost and loss in interpretability, an increase  in F1 score can justify such costs. The reason for this is that in this particular context, computational time is of a less concern. For a university administration having real-time predictions is very unlikely to be an important factor, whereas having higher student retention and higher student success rate in courses is a priority. Therefore, in my opinion it is safe to assume that IF ensembling can demonstrate substantially better results, then deploying it as a go-to model is a better choice.

For ensembling I will implement blending method. Blending is quite similar to stacking. The only difference is that training of meta-model is done on a separate holdout set (e.g 10% of training dataset) rather on full and folded training set. As parameters, I will pass first three ML algorithms that have shown the highest **F1** scores.

```python
### Ensembling
blended_model = blend_models(estimator_list = [lgbm_tuned, rf_tuned, gbc],
             choose_better = True,
             optimize = 'F1',             
             method = 'soft')
```

<br>

Unfortunately, ensembled model did not improve F1 score by a noticeable margin. Hence, it will not be deployed as a final model, since computational costs are not justified at all. 

Instead, for those tasks where **F1 score** will be used a <u>primary metric</u>,  **tuned version of Light GBM** will be used for model deployment. The logic behind this is that *Light GBM* demonstrated the same results that as *Random Forest* and *Gradient Boosting* algorithms, while having significantly lower computaional time.

### Setup 2 (ML Workflow for Maximizing Recall)

</br>

Now, let's build a prediction model for those courses (or even universities) that boast a high number of tutors relative to a class size. Given that those courses have relatively high amount of teaching resources, we can claim that **Recall**  is the most approopriate metric to use for such cases. The reason, as already mentioned, is that in this setup the cost of
*False Negatives* is significantly higher compared to the cost of *False Positives*. In other words, it is much worse to miss a failed student than to misclassify a successful student.

</br>

For this purpose, I will preprocess and feature engineer the data differently, with an intent to maximize **Recall** . Namely, I will add extra parameters like **feature interaction**  and **feature ratio**, since they seem to have a positive impact on our metric.
Since this will result in the creation of additional variables,  I will implement **Principal Component Analysis** in order to handle the problem known as *Curse of Dimensionality*. As we will see shortly, such transformation is indeed helpful in increasing **Recall** score by a large margin (unsurprisingly, at the cost of decrease in *Precision*). 

```python
df_setup_with_interaction = setup(data = data_for_modeling,
                 target = 'final_result', 
                 train_size = 0.3,
                 numeric_features = numeric_columns,  
                 categorical_features = categorical_columns,
                 fold = 5,    
                 fold_strategy = RepeatedStratifiedKFold(n_splits = 5, n_repeats = 3),
                 normalize = True, 
                 transformation_method = "yeo-johnson",
                 ignore_low_variance = True,
                 remove_multicollinearity = True,
                 multicollinearity_threshold = 0.95,
                 #create_clusters = True,
                 pca = True,                 
                 feature_interaction = True,
                 feature_ratio = True,
                 use_gpu = True,
                 silent = True, 
                 session_id = 123)
```

#### Comparison of Models

```python
models_comparison_with_interaction = compare_models(sort = 'Recall', round = 3, n_select = 3)
```

<br>

As seen from the comparison table above, **Quadratic Discriminant Analysis** alogirthm shows the highest **Recall** score. In plain English this means that out of all students who actually failed their Final Exams.

</br>

#### Hyperparameter Tuning


Despite the obtained results being promising, we still can try to increase the **Recall score** via tuning hyperparameters of the base model. Unlike tree-based models, Quadratic Discriminant Analysis algorithm does not have plenty of hyperparameters to tune. Still we can try and see if we can boost the performance. To do this, I will perform a randomized search with repeated and stratified cross validation.
